```python
# Here we will use some libraries of python to speed up the process
# nltk: Natural Language Toolkit, we can say nltk is a middle man between human and a machine to translate english to machine language
# pandas: pandas library is used to analyze the data
# numpy: Numpy is used to enhance the speed of mathematical operations, numpy is much faster than python's List

import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
import pandas as pd
import numpy as np 


```

```python
# Below code is just to read the dataset that is in .tsv format

dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SMSSpamCollection.tsv',sep = "\t",header = None)
dataset.columns = ['label','body_text']
dataset.head()
```

```python
# Below code is written to remove all the punctuation that are present in the given dataset

import string
string.punctuation 
def remove_punct(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    return text_nopunct

dataset['body_text_clean'] = dataset['body_text'].apply(lambda x: remove_punct(x))
dataset.head()
```

```python
# Below code is for making a new column for storing all the words in lowercase and also it gets stored individually like a list
import re

def tokenize(text):
    tokens = re.split('\W',text)
    return tokens

dataset['body_text_tokenized'] = dataset['body_text_clean'].apply(lambda x:tokenize(x.lower()))
dataset.head()
```

```python
# This code is used to remove stopwords from the dataset


stopwords = nltk.corpus.stopwords.words('english')

def remove_stopwords(tokenized_list):
    text=[word for word in tokenized_list if word not in stopwords]
    return text

dataset['body_text_nostop'] = dataset['body_text_tokenized'].apply(lambda x:remove_stopwords(x))

dataset.head()
```

```python
# below code is to remove 'ing' from the word and it is called as stemming

ps = nltk.PorterStemmer()

def stemming(tokenized_text):
    text = [ps.stem(word) for word in tokenized_text]
    return text

dataset['body_text_stemmed'] = dataset['body_text_nostop'].apply(lambda x:stemming(x))

dataset.head()
```

```python
# below code is used to group some words to analyzed as a single word is called lemmatization

wn = nltk.WordNetLemmatizer()

def lemmatizing(tokenized_text):
    text = [wn.lemmatize(word) for word in tokenized_text]
    return text

dataset['body_text_lemmatized'] = dataset['body_text_nostop'].apply(lambda x:lemmatizing(x))

dataset.head()
```

```python
#below code is used to convert text to a certain form where an system can understand what is it and it is called as  vectorization

from sklearn.feature_extraction.text import CountVectorizer
```

```python
# count vectorizer counts the number of words in dataset

def clean_text(text):
    text = "".join([char for char in text if char not in string.punctuation])
    tokens = re.split('\W',text)
    text = [ps.stem(word) for word in tokens if word not in stopwords]
    return text

count_vect = CountVectorizer(analyzer = clean_text)
X_count = count_vect.fit_transform(dataset['body_text'])
print(X_count.shape)
```

```python
# below code is use to show sparse matrix
#sparse matrix means where 1 is present those words will only gets stored to save memory 

x_counts_df = pd.DataFrame(X_count.toarray())
x_counts_df
```

```python
#below code is use to provide column name
x_counts_df.columns = count_vect.get_feature_names_out()
x_counts_df 
```

```python
# Same way of displaying the rows and columns, just we use sklearn library to show the rows and columns
#tfidf vectorization,where it gives the weightage where the spam word is present in that sentence
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
```

```python
tfidf_vect = TfidfVectorizer(analyzer = clean_text)
x_tfidf = tfidf_vect.fit_transform(dataset['body_text'])
print(x_tfidf.shape)

x_tfidf_df = pd.DataFrame(x_tfidf.toarray())
x_tfidf_df.columns = tfidf_vect.get_feature_names_out()
x_tfidf_df
```

```python
# Now we will do Feature engineering, it is used to manipulate the dataset to improve Machine Learning algorithm
# leading to greater Performance and accuracy

dataset = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/SMSSpamCollection.tsv",sep = "\t",header = None)
dataset.columns = ['label','body_text']
dataset.head()
```

```python
#count the percentage of punctuation present

def count_punct(text):
    count = sum([1 for char in text if char in string.punctuation])
    return round(count/(len(text)-text.count(" ")),3)*100

dataset["body_len"] = dataset["body_text"].apply(lambda x:len(x)-x.count(" "))

dataset["punct%"] = dataset["body_text"].apply(lambda x:count_punct(x))
dataset.head()
def clean_text(text):
    text = "".join([char for char in text if char not in string.punctuation])
    tokens = re.split('\W',text)
    text = [ps.stem(word) for word in tokens if word not in stopwords]
    return text

tfidf_vect = TfidfVectorizer(analyzer = clean_text)
x_tfidf = tfidf_vect.fit_transform(dataset['body_text'])
```

```python
#concatination (combine both dataset)

X_features = pd.concat([dataset['body_len'],dataset['punct%'],pd.DataFrame(x_tfidf.toarray())],axis=1)
X_features.head()
```

```python
# Model using train test split
# We have to use Random Forest Algorithm 

from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.model_selection import train_test_split
```

```python
# we have to split test and train data 

X_train, X_test, y_train, y_test = train_test_split(X_features, dataset['label'],test_size = 0.3, random_state = 0)
```

```python
# First we have to add data in train and test and then we have to train random forest model

rf= RandomForestClassifier(n_estimators=500, max_depth=20, n_jobs=-1,)
rf_model=rf.fit(X_train, y_train)
sorted(zip(rf_model.feature_importances_,X_train.columns),reverse=True)[0:10]

y_pred = rf_model.predict(X_test)

precision,recall,fscore,support = score(y_test, y_pred, pos_label = 'spam', average = 'binary')

print('Precision {} / Recall {} / Accuracy {}'.format(round(precision,3),
                                                      round(recall,3),round((y_pred==y_test).sum()/len(y_pred),3)))
```
